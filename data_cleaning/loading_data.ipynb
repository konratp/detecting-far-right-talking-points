# Loading Data

```python
import pandas as pd
import nltk
import os
```

```python
path = f"{os.getcwd()}/data_sample"
```

```python
os.listdir(path)
```

```python
corpus = {}
for i in os.listdir(path):
    corpus[i] = nltk.data.load(f'{path}/{i}', format='raw')             
```

```python
#corpus2 = pd.DataFrame.from_dict(corpus, index = [0])
corpus2 = pd.DataFrame.from_dict([corpus])
```

```python
corpus2 = corpus2.T
```
