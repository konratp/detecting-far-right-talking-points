```python
spea_split = []
for i in chap_split:
    spea_split.append(i.split(b"SPEA"))
```

```python
for i in spea_split:
    print(len(i))
```

```python
test_corp = pd.DataFrame(spea_split).T
test_corp
```

```python
d = {}
counter = 0
for text in spea_split:
    d[counter] = []
    for contribution in text:
        d[counter].append(contribution)
    counter += 1
```

```python
d[4]
```

```python
entries = []
chapter = []
for i in d:
    entries.extend(d[i])
    for k in range(len(d[i])):
        chapter.append(i)
```

```python
titles = []
for i in d:
    titles.append(d[i][0])
    
titles   
```

```python
results = pd.DataFrame({'contribution': entries,'chapter': chapter})
results
```
